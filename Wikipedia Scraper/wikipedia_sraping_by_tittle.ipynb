```python
import requests
from bs4 import BeautifulSoup
```

```python
# Get the topic input from the user and replace spaces with '+' for URL formatting
title = str(input('Enter the topic: ')).replace(' ', '+')

# Construct the Google search URL for the topic, specifically targeting Wikipedia
link = 'https://www.google.com/search?q=' + title + '+wikipedia'

# Send a GET request to the constructed URL and store the response
res = requests.get(link)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')

# Loop through all 'div' elements in the parsed HTML
for sp in soup.find_all('div'):
    try:
        # Attempt to find the 'a' tag and get its 'href' attribute (the link)
        link = sp.find('a').get('href')
        
        # Check if the link contains 'en.wikipedia.org', indicating it's a Wikipedia link
        if ('en.wikipedia.org' in link):
            break  # Exit the loop if a Wikipedia link is found
    except:
        # If there's an error (e.g., 'a' tag doesn't exist), just continue to the next 'div'
        pass

# Clean up the link by removing Google's tracking and unnecessary parameters
link = link[7:].split('&')[0]
```

```python
# Send a GET request to the Wikipedia link and store the response
res = requests.get(link)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')

# Initialize an empty string to store the extracted text
corpus = ''

# Loop through all 'p' (paragraph) elements in the parsed HTML
for p in soup.find_all('p'):
    corpus += p.text  # Append the text from each paragraph to the corpus
    corpus += '\n'    # Add a newline character for readability

# Remove leading and trailing whitespace from the corpus
corpus = corpus.strip()

# Remove citation references (e.g., [1], [2], etc.) from the corpus text
for i in range(500):  # Assumes a max of 500 citations
    corpus = corpus.replace('[' + str(i) + ']', '')
```
